# Frecuencia de anotaciones (estado E9)

Este Script permite la realización de un histograma para visualizar la frecuencia de cada una de las anotaciones asociadas a los segmentos de interés (segmentos de 200 pares de bares, con estado E9 y filtrados)

```python
# Importe de los módulos necesarios
import csv
import matplotlib.pyplot as plt

# Se carga el archivo que contiene la frecuencia de cada anotación y se introduce en un diccionario:
#    Claves -> anotaciones
#    Valores -> frecuencia por anotación
reader = csv
with open('./ARCHIVOS_PARA_GRAFICA/Etiquetas_Valores_anotaciones.csv', mode='r') as f:
    reader = csv.reader(f)
    diccionario_anotaciones = {rows[0]:rows[1] for rows in reader}
```

```python
# Cambio de los valores del diccionario strings a integers.
for key in diccionario_anotaciones:
    diccionario_anotaciones[key] = int(diccionario_anotaciones[key])
```

```python
# Modificación del nombre de las claves del diccionario para clarificar las etiquetas del histograma
diccionario_anotaciones['InterCGI'] = diccionario_anotaciones.pop('hg19_cpg_inter')
diccionario_anotaciones['CpG Islands'] = diccionario_anotaciones.pop('hg19_cpg_islands')
diccionario_anotaciones['CpG shelves'] = diccionario_anotaciones.pop('hg19_cpg_shelves')
diccionario_anotaciones['CpG shores'] = diccionario_anotaciones.pop('hg19_cpg_shores')
diccionario_anotaciones['1to5kb'] = diccionario_anotaciones.pop('hg19_genes_1to5kb')
diccionario_anotaciones['3UTRs'] = diccionario_anotaciones.pop('hg19_genes_3UTRs')
diccionario_anotaciones['5UTRs'] = diccionario_anotaciones.pop('hg19_genes_5UTRs')
diccionario_anotaciones['CDS'] = diccionario_anotaciones.pop('hg19_genes_cds')
diccionario_anotaciones['Exons'] = diccionario_anotaciones.pop('hg19_genes_exons')
diccionario_anotaciones['Intergenic'] = diccionario_anotaciones.pop('hg19_genes_intergenic')
diccionario_anotaciones['Intron/Exon boundaries'] = diccionario_anotaciones.pop('hg19_genes_intronexonboundaries')
diccionario_anotaciones['Introns'] = diccionario_anotaciones.pop('hg19_genes_introns')
diccionario_anotaciones['Promoters'] = diccionario_anotaciones.pop('hg19_genes_promoters')
```

```python
# Asignación de las claves y los valores a dos listas
lista_claves = list(diccionario_anotaciones.keys())
lista_valores = list(diccionario_anotaciones.values())

# Paso de los valores a porcentajes para poder realizar la comparativa con todo el genoma
lista_porcentajes = []
total_anotaciones = sum(lista_valores)
for i in lista_valores:
    lista_porcentajes.append(i/sum(lista_valores))
```

```python
# Generación de la gráfica de visualización
plt.figure(figsize=(10,10))
plt.xticks(rotation=45, ha="right")
plt.ylim(ymax = 0.40, ymin = 0)
plt.ylabel('Frecuencia por anotación', fontsize = 15)
plt.xlabel('Anotaciones', fontsize = 15)
plt.title('Frecuencia de cada una de las anotaciones (Estado 9)', fontsize = 20, y = 1.03)
plt.bar(lista_claves,lista_porcentajes, color=['seagreen','seagreen','seagreen','seagreen','slateblue', \
    'mediumorchid','mediumorchid','aquamarine','aquamarine','coral','coral','coral','darkgoldenrod'])
plt.show()
```
